### Import Pustaka

Untuk analisa ini saya akan menggunakan pandas dan numpy sebagai library utama

### Data Loading

Nama file yang akan digunakan ialah "Pokemon.csv"

```python
stats = pd.read_csv('Pokemon.csv')
```

### Data Cleaning

Untuk menjaga agar data yang asli tetap original dan tidak berubah, maka data tersebut akan di salinkan terlebih dahulu dengan nama variabel "stats_clear"

```python
stats_clear = stats.copy()
```

Pada data terlihat ada "NaN". Value ini dapat diganti menjadi "-" agar lebih mudah dibaca


```python
stats_clear.fillna('-')
```

Beberapa kolom yang menurut saya tidak dibutuhkan pada analisa saat ini dapat dihapus dengan menggunakan command drop


```python
gen = stats_clear.drop(['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'], axis = 1)
```

### Eksplorasi Data

Nilai mean adalah nilai rata-rata, maka dari itu saya akan mencari tahu nilai mean pada "Total" melalui command describe

```python
stats_clear.describe()
```

Pada script ini saya mencoba untuk menggunakan data querying dan grouping untuk mengeluarkan data-data yang hendak di analisa dari data originalnya.

Telah diketahui bahwa nilai mean dari "Total" adalah 435, jadi saya akan menganggap bahwa nilai dibawah 435 sebagai angka dibawah rata-rata dan nilai diatas 435 sebagai angka diatas rata-rata

```python
gen1_stats_high = stats_clear.query('Total > 435 and Generation == 1')
gen1_stats_low = stats_clear.query('Total < 435 and Generation == 1')

gen2_stats_high = stats_clear.query('Total > 435 and Generation == 2')
gen2_stats_low = stats_clear.query('Total < 435 and Generation == 2')

gen3_stats_high = stats_clear.query('Total > 435 and Generation == 3')
gen3_stats_low = stats_clear.query('Total < 435 and Generation == 3')

gen4_stats_high = stats_clear.query('Total > 435 and Generation == 4')
gen4_stats_low = stats_clear.query('Total < 435 and Generation == 4')

gen5_stats_high = stats_clear.query('Total > 435 and Generation == 5')
gen5_stats_low = stats_clear.query('Total < 435 and Generation == 5')

gen6_stats_high = stats_clear.query('Total > 435 and Generation == 6')
gen6_stats_low = stats_clear.query('Total < 435 and Generation == 6')
```

Kemudian data-data tersebut saya urutkan tabelnya agar menjadi lebih mudah untuk dibaca dan dipahami

```python
# Data di sort agar lebih mudah dibaca
gen1_stats_high = gen1_stats_high.sort_values(by=['Total'])
gen1_stats_low = gen1_stats_low.sort_values(by=['Total'])

gen2_stats_high = gen2_stats_high.sort_values(by=['Total'])
gen2_stats_low = gen2_stats_low.sort_values(by=['Total'])

gen3_stats_high = gen3_stats_high.sort_values(by=['Total'])
gen3_stats_low = gen3_stats_low.sort_values(by=['Total'])

gen4_stats_high = gen4_stats_high.sort_values(by=['Total'])
gen4_stats_low = gen4_stats_low.sort_values(by=['Total'])

gen5_stats_high = gen5_stats_high.sort_values(by=['Total'])
gen5_stats_low = gen5_stats_low.sort_values(by=['Total'])

gen6_stats_high = gen6_stats_high.sort_values(by=['Total'])
gen6_stats_low = gen6_stats_low.sort_values(by=['Total'])
```
