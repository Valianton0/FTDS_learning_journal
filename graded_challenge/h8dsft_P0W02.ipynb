Berikut ini untuk menjalankan fungsi gradien untuk masing-masing sumbu x dan sumbu y pada matriks gambar

```python
def f(x,y):
    return x**2*y+2*x**3*y+y**4

x = np.linspace(438,780)
y = np.linspace(438,780)

z = np.array([[f(i,j) for i in x] for j in y])

dx, dy = np.gradient(z)
```

Selanjutnya kita akan menghitung gradient magnitudenya dengan bantuan scipy

```python
from scipy import ndimage

dx = ndimage.sobel(greyimg2, axis=0, mode='constant')

dy = ndimage.sobel(greyimg2, axis=1, mode='constant')

dxy = np.hypot(dx,dy)
```

### HASIL DAN KESIMPULAN

Pada script ini saya mencoba mencari edge terbaik pada gambar. Dengan pemahaman saya saat ini threshold terbaik untuk melihat edge pada gambar tersebut adalah 235.

```python
threshold = 235

edge = np.where(dxy > threshold, 255, 0)

plt.imshow(edge, cmap=plt.cm.gray)
plt.show()
```
